Created tables in new schema

```python
%sql
CREATE OR REPLACE TABLE bronze.events
AS
SELECT * FROM delta.`/Volumes/ecom/bronze/ecom_data/ecom_bronze_events_delta`;

CREATE OR REPLACE TABLE silver.events
AS
SELECT * FROM delta.`/Volumes/ecom/silver/ecom_silver/ecom_silver_events_delta`;

CREATE OR REPLACE TABLE gold.viewed
AS
SELECT * FROM delta.`/Volumes/ecom/gold/ecom_gold/ecom_gold_user_viewed`;

CREATE OR REPLACE TABLE gold.cart
AS
SELECT * FROM delta.`/Volumes/ecom/gold/ecom_gold/ecom_gold_user_cart`;
```
